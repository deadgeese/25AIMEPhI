```python
def even_reverse(lst):
    n = len(lst)
    mid = n // 2
    even_elements = lst[mid::2]
    reversed_even = even_elements[::-1]
    lst[mid::2] = reversed_even
    return lst

lst = [1, 45, 8, 123, 89]
print(even_reverse(lst))
```

```python
def swap_blocks(lst):
    temp = lst[2:5:2].copy()
    lst[2:5:2] = lst[-4:-1:2]
    lst[-4:-1:2] = temp
    return lst
lst = np.arange(1, 21)
print(lst)
print(swap_blocks(lst))

```

```python
# Прямые заданы уравнением Ax + By + C = 0
lines = [
    [1, 0, -2],
    [0, 1, -3],
    [1, -1, 0]
]
point = (1, 1)

distances = []
for A, B, C in lines:
    distance = abs(A*point[0] + B*point[1] + C) / np.sqrt(A**2 + B**2)
    distances.append(distance)

closest_line_index = np.argmin(distances)
print(f"Ближайшая прямая: {lines[closest_line_index]}, расстояние: {distances[closest_line_index]}")
```

```python
numeric_cols = df.select_dtypes(include=[np.number]).columns
plt.figure(figsize=(15, 30))
for i, column in enumerate(numeric_cols, 1):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outliers = df.loc[(df[column] < (Q1 - 1.5 * IQR)) | (df[column] > (Q3 + 1.5 * IQR))].dropna(how='all')
    print(outliers)
    plt.subplot(5, 4, i)
    sns.boxplot(y=df[column].dropna())
    plt.title(f'Ящик с усами для {column}')
    plt.tight_layout()
plt.show()
```

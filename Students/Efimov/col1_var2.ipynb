
- Постройте boxplot для всех переменных и выведете объекты которые являются выбросами (строки из датафрейма).

```python
df.plot.box(subplots=True, figsize=(15, 10), layout=(3, 4))
```

```python
def detect_outliers_iqr(df, column):
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    interval = 1.5 * (Q3 - Q1)
    l, r = Q1 - interval, Q3 + interval
    return df[(df[column] < l) | (r < df[column])]

for c in df.columns:
    if pd.api.types.is_numeric_dtype(df[c]):
        temp = detect_outliers_iqr(df, column=c)
        print(f"There are {temp.shape[0]} outliers for {c}")
        display(temp.head(15))
```

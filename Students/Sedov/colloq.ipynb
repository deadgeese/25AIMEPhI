## NumPy

```python
import numpy as np
from numpy.lib.stride_tricks import sliding_window_view
```

Дан массив 16х16
Создайте новый массив 4х4 , элементы которого будут суммой блоков 4х4 исходного массива

```python
orig = np.arange(16*16).reshape(16, 16)
print(orig)
sliding_window_view(orig, (4, 4)).sum(axis=(-2, -1))[::4, ::4]
```

Дан массив, вставьте между каждым значением сумму соседних элементов

```python
array = np.arange(5)
res = np.zeros(2*array.shape[0] - 1)
res[::2] = array
res[1:-1:2] = array[:-1] + array[1:]
res
```

### Nans

```python
print(f"total number of nans in each column:\n{df.isna().sum()}")
sns.heatmap(df.isna().transpose(), cbar_kws={'label': 'Missing Data'})
plt.show()
```

```python
print(f"Ratio of nans in each column:\n{df.isna().sum() / df.shape[0]}")
```

### Actual productivity distribution

```python
fig, ax = plt.subplots(1, 2, figsize=(10, 5))
sns.histplot(df.actual_productivity, kde=True, ax=ax[0])
sns.boxplot(df.actual_productivity, ax=ax[1], orient='h')
plt.show()
```

### Correlation

```python
fig, ax = plt.subplots(figsize=(10, 10))
corr = df.select_dtypes(include='number').corr()
mask = np.triu(np.ones_like(corr), k=1)
np.triu(np.ones_like(corr))
sns.heatmap(corr, annot=True, ax=ax, mask=mask)
plt.show()
```

### Pairplot

### Boxplots and outliers

```python
nrows = 3
ncols = 3
fig, ax = plt.subplots(nrows, ncols,figsize=(10, 6))
```

```python
outlier_mask = np.zeros(df.shape[0], dtype=bool)
for column in df.select_dtypes(include='number').columns:
    Q1 = df[column].quantile(0.25)
    Q3 = df[column].quantile(0.75)
    IQR = Q3 - Q1
    outlier_mask[(df[column] < Q1 - 1.5 * IQR) | (df[column] > Q3 + 1.5 * IQR)] = True
```

```python
df[outlier_mask]
```

### Quarters

```python
target_columns = set(df.select_dtypes(include=['number']).columns) - set(['team'])
target_columns.add('quarter')
target_columns = list(target_columns)
```

#### Cписки

Вам дан список. Поменяйте порядок следования каждого чётного элемента во второй половине списка.

```python
def reverse_even_in_second_half_v2(lst):
    n = len(lst)
    mid = n // 2
    
    second_half = lst[mid:].copy()
    
    even_indices = [i for i, x in enumerate(second_half) if x % 2 == 0]
    even_values = [second_half[i] for i in even_indices]
    
    reversed_evens = even_values[::-1]
    for idx, new_val in zip(even_indices, reversed_evens):
        second_half[idx] = new_val
    
    return lst[:mid] + second_half

input_list = [1, 2, 3, 4, 5, 6, 7, 8, 12, 14]
output_list = reverse_even_in_second_half_v2(input_list)
print("Исходный список:", input_list)
print("Результат:", output_list)
```

```python
def swap_blocks_v2(lst):
    n = len(lst)
    

    block_A_indices = [i for i in range(1, min(5, n))]
    block_B_indices = [n-i for i in range(2, 6) if n-i >= 0]
    
    even_A = [(i, lst[i]) for i in block_A_indices if lst[i] % 2 == 0]
    odd_B = [(i, lst[i]) for i in block_B_indices if lst[i] % 2 == 1]
    
    for (idx_a, _), (idx_b, _) in zip(even_A, odd_B):
        lst[idx_a], lst[idx_b] = lst[idx_b], lst[idx_a]
    
    return lst

# Тестирование альтернативной версии
print("\nАльтернативная версия:")
test_list = [0, 1, 2, 3, 4, 45, 54, 100, 5, 6, 7, 8, 9]
print(f"Исходный: {test_list}")
print(f"Результат: {swap_blocks_v2(test_list)}")
```

```python
def decrease_elements(lst, dcr):
    return [x - dcr for x in lst]

# Пример использования первого задания
test_list1 = [5, 10, 15, 20]
result1 = decrease_elements(test_list1, 2)
print(f"Исходный список: {test_list1}")
print(f"Элементы уменьшены на 1: {result1}")
print()
```

```python
def process_even_odd(lst):
    return [x**2 if x % 2 == 0 else x**3 for x in lst]

# Пример использования второго задания
test_list2 = [1, 2, 3, 4, 5, 6]
result2 = process_even_odd(test_list2)
print(f"Исходный список: {test_list2}")
print(f"Четные в квадрат, нечетные в куб: {result2}")
print()
```

#### Встроенные функции

```python
def min_ascii_code(s):
    if not s:  
        return None
    
    min_char = min(s)  
    ascii_code = ord(min_char)
    return ascii_code

# Пример использования 
test_string1 = "Hello World!"
result1 = min_ascii_code(test_string1)
print(f"Строка: '{test_string1}'")
print(f"Минимальный символ: '{min(test_string1)}'")
print(f"Его ASCII код: {result1}")
print()
```

```python
def shift_chars_by_index(s):
    result_chars = []
    for i, char in enumerate(s):
        new_ascii = ord(char) + i
        new_char = chr(new_ascii)
        result_chars.append(new_char)
    
    return ''.join(result_chars)

# Пример использования 
test_string2 = "abc"
result2 = shift_chars_by_index(test_string2)
print(f"Исходная строка: '{test_string2}'")
print(f"После смещения: '{result2}'")
```

```python
def sum_4x4_blocks(arr_16x16):
    if arr_16x16.shape != (16, 16):
        raise ValueError("Массив должен быть размером 16x16")
    
    result = arr_16x16.reshape(4, 4, 4, 4).sum(axis=(1, 3))
    return result

print("Первое задание:")
np.random.seed(42)
original_array = np.random.randint(0, 2, (16, 16))
print("Исходный массив 16x16:")
print(original_array)
print("\nРезультирующий массив 4x4 (суммы блоков 4x4):")
result_array = sum_4x4_blocks(original_array)
print(result_array)
# print(original_array.reshape(4, 4, 4, 4)[0, 0])
```

```python
def distance_to_lines(point, lines):
    x0, y0 = point
    distances = []
    
    for A, B, C in lines:
        # Формула расстояния от точки до прямой: |Ax0 + By0 + C| / sqrt(A^2 + B^2)
        numerator = abs(A * x0 + B * y0 + C)
        denominator = np.sqrt(A**2 + B**2)
        distance = numerator / denominator
        distances.append(distance)
    
    distances = np.array(distances)
    min_distance = np.min(distances)
    closest_line_index = np.argmin(distances)
    
    return min_distance, closest_line_index, distances

# Тестирование 
print("\nВторое задание:")

point = (2, 3)
lines = [
    (1, -1, 0),   # x - y = 0
    (2, 3, -5),   # 2x + 3y - 5 = 0
    (1, 0, -4),   # x - 4 = 0 (вертикальная)
    (0, 1, -2),   # y - 2 = 0 (горизонтальная)
    (3, -2, 1)    # 3x - 2y + 1 = 0
]

min_dist, closest_idx, all_distances = distance_to_lines(point, lines)

print(f"Точка: {point}")
print("Прямые (Ax + By + C = 0):")
for i, (A, B, C) in enumerate(lines):
    print(f"  {i}: {A}x + {B}y + {C} = 0")
print(f"\nРасстояния до прямых: {all_distances}")
print(f"Ближайшая прямая: №{closest_idx} ({lines[closest_idx]})")
print(f"Минимальное расстояние: {min_dist:.4f}")
```

```python
# Boxplot для числовых переменных
plt.figure(figsize=(15, 8))
df[numeric_columns].boxplot()
plt.title('Boxplot всех числовых переменных')
plt.xticks(rotation=45)
plt.tight_layout()
plt.show()

def find_outliers_iqr(column):
    Q1 = column.quantile(0.25)
    Q3 = column.quantile(0.75)
    IQR = Q3 - Q1
    lower_bound = Q1 - 1.5 * IQR
    upper_bound = Q3 + 1.5 * IQR
    return column[(column < lower_bound) | (column > upper_bound)]

outlier_rows = set()
for column in numeric_columns:
    outliers = find_outliers_iqr(df[column])
    outlier_rows.update(outliers.index)

print(f"Количество строк с выбросами: {len(outlier_rows)}")
if outlier_rows:
    print("Пример строк с выбросами:")
    display(df.loc[list(outlier_rows)[:10]])
```

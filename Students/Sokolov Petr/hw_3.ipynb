```python
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('train_data.csv')
df
```

```python
df.info()
```

```python
df_nan = df.loc[:, df.isnull().any()]
df_nan
```

```python
df_nan.isnull().sum() / len(df_nan) * 100
```

```python
df = df.drop(['Domains', 'Mechanics', 'ID', 'Name'], axis=1)
```

```python
df['Owned Users'].fillna(df['Owned Users'].mean(), inplace=True)
```

```python

df = df.dropna(axis=0)
df.info()
```

```python
df['Complexity Average'] = df['Complexity Average'].replace(',', '.', regex=True).astype(float)
df['Rating Average'] = df['Rating Average'].replace(',', '.', regex=True).astype(float)
df.describe()
```

```python
df.loc[df['Min Players'] == 0, 'Min Players'] = 1
df.loc[df['Max Players'] == 0, 'Max Players'] = df['Min Players']

df[df['Min Players'] == 0]

# df.loc[df['Play Time'] == 0, 'Play Time'] = 1
df.describe()
```

```python
X = df.drop(['Rating Average'], axis=1)
Y = df['Rating Average']
```

```python

from sklearn.preprocessing import MinMaxScaler, StandardScaler

std = StandardScaler()

X_std = std.fit_transform(X) # np array

X_std = pd.DataFrame(X_std, columns=X.columns)
X_std
```

```python
X_std.describe()
```

```python
mmsc = MinMaxScaler()
X_norm = mmsc.fit_transform(X)

X_norm = pd.DataFrame(X_norm, columns=X.columns)

X_norm.describe()
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_norm, Y, test_size=0.15, random_state=42)

print(f'Train dataset size: {X_train.shape}, {y_train.shape}')
print(f'Train dataset size: {X_test.shape}, {y_test.shape}')
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)
```

```python
y_pred = model.predict(X_test)
print(y_pred)
```

```python
from sklearn import metrics

print("MAE", metrics.mean_absolute_error(y_test, y_pred))
print("MSE", metrics.mean_squared_error(y_test, y_pred))
print("R2 Score:", model.score(X_test, y_test))
```

```python
features = X.columns

coeff_df = pd.DataFrame(model.coef_, columns=['Coefficient'])
coeff_df['features'] = features

coeff_df.sort_values(by='Coefficient')
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_norm, Y, test_size=0.15, random_state=42)

model = LinearRegression()

model.fit(X_train, y_train)
y_pred = model.predict(X_test)

print("MAE", metrics.mean_absolute_error(y_test, y_pred))
print("MSE", metrics.mean_squared_error(y_test, y_pred))
print("Score:", model.score(X_test, y_test))
```

```python
from sklearn.model_selection import KFold, cross_val_score

# Define the number of folds for cross-validation
k = 5
kf = KFold(n_splits=k, shuffle=True, random_state=42)
```

```python
lin_reg = LinearRegression()

scores = cross_val_score(lin_reg, X, Y, cv=kf, scoring='r2')
scores
```

```python
mean_score = np.mean(scores)
mean_score
```

## Гипотезы

**Доверительный интервал для среднего**

Пусть дана выборка данных $(x_1, x_2, \dots, x_n)$.  
Необходимо оценить среднее значение $\bar{x}$ и построить доверительный интервал с уровнем доверия $\text{confidence}$.

Стандартная ошибка среднего определяется как:
$$
SE = \frac{s}{\sqrt{n}}
$$
где $s$ — стандартное отклонение выборки, а $n$ — размер выборки.

Половина ширины доверительного интервала вычисляется по формуле:
$$
h = SE \cdot t_{(1 + \text{confidence})/2, \, n-1}
$$
где $t_{(1 + \text{confidence})/2, \, n-1}$ — квантиль t-распределения Стьюдента с $n-1$ степенями свободы.

Доверительный интервал для среднего:
$$
\text{CI} = [\bar{x} - h, \, \bar{x} + h]
$$

**Коэффициент эффекта Cohen's d**

Пусть есть две выборки $X = \{x_1, \dots, x_{n_X}\}$ и $Y = \{y_1, \dots, y_{n_Y}\}$.  
Cohen's d измеряет величину различия между средними в терминах стандартного отклонения.                                                                     Cohen's d вычисляется как:
$$
d = \frac{\bar{X} - \bar{Y}}{s_p}
$$

Интерпретация величины эффекта:
$$
d \approx
\begin{cases}
0.2 & \text{малый эффект} \\
0.5 & \text{средний эффект} \\
0.8 & \text{большой эффект}
\end{cases}
$$

**Значение t-статистики**

Пусть есть две независимые выборки $X_1$ и $X_2$ с размерами $n_1$ и $n_2$, средними $\bar{X}_1$, $\bar{X}_2$ и дисперсиями $s_1^2$, $s_2^2$.  

t-статистика (для Welch’s t-test) вычисляется как:
$$
t = \frac{\bar{X}_1 - \bar{X}_2}{\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}}
$$

- Числитель $\bar{X}_1 - \bar{X}_2$ — **абсолютная разница средних**.  
- Знаменатель $\sqrt{\frac{s_1^2}{n_1} + \frac{s_2^2}{n_2}}$ — **стандартная ошибка разницы средних**.  

**Интерпретация:**

- $t$ показывает, **сколько стандартных ошибок помещается в наблюдаемую разницу средних**.  
- Например, если $t = 2$, это означает, что разница средних равна примерно двум стандартным ошибкам.  
- Знак $t$ (+ или -) показывает **направление различия** между выборками.  

t-статистика — это **относительное значение**, нормированное на стандартную ошибку


```python
from scipy import stats

def mean_confidence_interval(data, confidence=0.95):
    a = np.array(data.dropna())
    n = len(a)
    if n == 0:
        return (np.nan, np.nan, n)
    m = a.mean()
    se = stats.sem(a)
    h = se * stats.t.ppf((1 + confidence) / 2., n-1)
    return (m - h, m + h, n)

def cohens_d(x, y):
    x = np.asarray(x.dropna()); y = np.asarray(y.dropna())
    nx, ny = len(x), len(y)
    if nx < 2 or ny < 2:
        return np.nan
    dof = nx + ny - 2
    pooled_var = ((nx - 1) * x.var(ddof=1) + (ny - 1) * y.var(ddof=1)) / dof
    return (x.mean() - y.mean()) / np.sqrt(pooled_var)
```

Гипотеза 1

Силовые тренировки эффективнее остальных видов тренировок.

```python
df['is_strength'] = df['Workout_Type'].str.contains('Strength', case=False, na=False)

strength = df[df['is_strength']]['Calories_Burned'].dropna()
non_strength = df[~df['is_strength']]['Calories_Burned'].dropna()
```

```python
t_stat, p_val = stats.ttest_ind(strength, non_strength, equal_var=False)

strength_ci = mean_confidence_interval(strength)
non_strength_ci = mean_confidence_interval(non_strength)

d = cohens_d(strength, non_strength)

print(f"Среднее силовые = {strength.mean():.1f} калорий, 95% CI = ({strength_ci[0]:.1f}, {strength_ci[1]:.1f}), n={strength_ci[2]}")
print(f"Среднее не-силовые = {non_strength.mean():.1f} калорий, 95% CI = ({non_strength_ci[0]:.1f}, {non_strength_ci[1]:.1f}), n={non_strength_ci[2]}")
print(f"t = {t_stat:.3f}, p = {p_val:.3e}, Cohen's d = {d:.3f}")
```

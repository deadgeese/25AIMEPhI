```python
arr = list(input().split())
mid = int(len(arr) / 2)
first_half = arr[: mid: ].copy()
second_half = arr[mid: : ].copy()

if mid % 2 == 0:
  start = 0
else:
  start = 1

temp = 0

for i in range(start, int(len(second_half) / 2), 2):
    temp = second_half[i]
    second_half[i] = second_half[-i - 1]
    second_half[-i - 1] = temp

print(first_half + second_half)
```

```python
arr = list(map(int, input().split()))
new_arr = []

for i in range (len(arr)):
  if i % 2 == 0:
    new_arr.extend([arr[i] ** 2])
  else:
    new_arr.extend([arr[i] ** 3])

print(new_arr)
```

```python
word = str(input())
word_dict = dict()

for symbol in word:
  word_dict[symbol] = ord(symbol)

print(min(word_dict, key=word_dict.get))
```

```python
word = str(input())
new_word = ''

for ind, symbol in enumerate(word):
   new_word += chr(ord(symbol) + ind)

print(new_word)
```

```python
import numpy as np

arr_16 = np.arange(256).reshape(16, 16)

arr_4 = np.zeros((4, 4), dtype=int)

for i in range(4):
    for j in range(4):
        submatrix = arr_16[i*4:(i+1)*4, j*4:(j+1)*4]
        arr_4[i, j] = np.sum(submatrix)

print(arr_4)
```

```python
import numpy as np

lines = np.array([[2, -1, 3],
                  [1, 1, -5],
                  [-3, 4, 2],
                  [0, 1, -2]])

point = np.array([6, 7])

dist = []

for i, line in enumerate(lines):
  first = np.abs(line[0] * point [0] + line[1] * point [1] + line[2])
  second = np.sqrt(line[0] ** 2 + line[1] ** 2)
  dist.insert(i, first / second)

dist = np.array(dist)

print(f"""минимальное расстояние {dist.min()}
прямая с минимальным расстоянием {lines[np.argmin(dist)]}""")
```

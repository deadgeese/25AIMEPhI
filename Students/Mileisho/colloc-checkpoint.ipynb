```python
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import pandas as pd

dt = pd.read_csv('Train.csv')

print("data types:")
print(dt.dtypes)
```

```python
print("missed values:")
missed_val = dt.isnull().sum()
print(missed_val)

plt.figure(figsize=(12, 5))
cmap = sns.cubehelix_palette(as_cmap=True, light=.9)
sns.heatmap(dt.isna().transpose(), cmap=cmap,
            cbar_kws={'label': 'missed  values'})
```

```python
dt.describe()
```

```python
plt.figure(figsize=(12, 5))

plt.hist(dt['Attrition_rate'], bins=30, alpha=0.7, color='orange', edgecolor='black')
plt.xlabel('Attrition Rate')
plt.ylabel('number')

plt.show()
```

```python
print("heatmap")
num_columns = dt.select_dtypes(include=[np.number]).columns

plt.figure(figsize=(15, 12))
correlation = dt[num_columns].corr()
sns.heatmap(correlation, cmap='coolwarm', center=0, annot=True)
plt.show()
```

```python
sns.pairplot(dt)
plt.show()
```

```python
num_columns = dt.select_dtypes(include=np.number).columns

for col in num_columns:
    plt.figure(figsize=(8,5))
    sns.boxplot(y=dt[col])
    plt.title(f"boxplot for {col}")
    plt.show()

    Q1 = dt[col].quantile(0.25)
    Q3 = dt[col].quantile(0.75)
    IQR = Q3 - Q1
    outliers = dt[(dt[col] < Q1 - 1.5*IQR) | (dt[col] > Q3 + 1.5*IQR)]
    if not outliers.empty:
        print(f"throws for {col}:")
        print(outliers)
        print("\n")
```

```python
print("relation status")
if 'Relationship_Status' in dt.columns:
    stats_by_relationship = dt.groupby('Relationship_Status').agg({
        'Age': ['avg', 'median', 'min', 'max'],
        'Attrition_rate': ['avg', 'median', 'min', 'max'],
        'Pay_Scale': ['avg', 'median', 'min', 'max'],
        'Time_of_service': ['avg', 'median', 'min', 'max']
    }).round(2)
    print(stats_by_relationship)
```

```python
#8
if 'Unit' in dt.columns:
    plt.figure(figsize=(12, 6))
    dt['Unit'].value_counts().plot(kind='bar')
    plt.title('number in Unit/Department')
    plt.xlabel('Unit')
    plt.ylabel('number of departments')
    plt.xticks(rotation=45)
    plt.tight_layout()
    plt.show()
    
    most_popular_unit = dt['Unit'].value_counts().index[0]
    print(f"the popular department: {most_popular_unit}")
    print(f"number of employees: {dt['Unit'].value_counts().iloc[0]}")

#9
average_attrition = dt['Attrition_rate'].mean()
mask = (dt['Unit'] == most_popular_unit) & (dt['Attrition_rate'] > average_attrition) & (dt['Relationship_Status'] == 'Married')
count_married_high_attrition = dt[mask].shape[0]
print(f"number of married with Attrition_rate above average in {most_popular_unit}: {count_married_high_attrition}")     
```

```python
plt.figure(figsize=(8, 6))

platform_crosstab = pd.crosstab(data['platform'], data['is_misinformation'], normalize='index')

platform_crosstab.columns = ['truth', 'misinformation']

platform_crosstab.plot(kind='bar', 
                      color=['#2ecc71', '#e74c3c'],
                      figsize=(8, 6))
plt.title('parts of truth and misinformation on platforms', fontsize=16, fontweight='bold')
plt.xlabel('platform', fontsize=12)
plt.ylabel('parts', fontsize=12)
plt.legend(title='content type')
plt.xticks(rotation=45)
plt.grid(axis='y', alpha=0.3)
plt.tight_layout()
plt.show()

print("parts on platforms:")
print(platform_crosstab.round(3))
```

```python
plt.figure(figsize=(15, 10))
cols = ['author_followers', 'text_length', 'readability_score', 
                'num_mentions', 'sentiment_score', 
                'toxicity_score', 'detected_synthetic_score', 'embedding_sim_to_facts',
                'external_factchecks_count', 'source_domain_reliability',
                'is_misinformation']
corr_matrix = data[cols].corr()
sns.heatmap(corr_matrix, annot=True, cmap='coolwarm', center=0)
plt.title('correlation')
plt.tight_layout()
plt.show()
```

```python
fig, (ax1, ax2) = plt.subplots(1, 2, figsize=(18, 6))

month_misinfo = pd.crosstab(data['month'], data['is_misinformation'])[1]
month_misinfo.sort_index().plot(kind='line', marker='s', ax=ax1, title='misinformation by monthes')
ax1.grid(True, alpha=1)

weekday_misinfo = pd.crosstab(data['weekday'], data['is_misinformation'])[1]
weekday_misinfo.sort_index().plot(kind='line', marker='s', ax=ax2, title='misinformation by week days')
ax2.grid(True, alpha=1)
plt.tight_layout()
plt.show()

print("\n1. misinformation by week days:")
for day, count in weekday_misinfo.items():
    print(f"   {day}: {count}")

print("\n2. misinformation by monthes:")
for month, count in month_misinfo.items():
    print(f"   {month}: {count}")
```

```python
import pandas as pd
```

```python
df = pd.read_csv('train_data.csv')
df
```

```python
df.info()
```

```python
df["Rating Average"] = pd.to_numeric(
    df["Rating Average"].astype(str).str.replace(',', '.'), 
    errors='coerce'
)

df["Complexity Average"] = pd.to_numeric(
    df["Complexity Average"].astype(str).str.replace(',', '.'), 
    errors='coerce'
)
df.info()
```

```python
df_nan = df.loc[:, df.isnull().any()]
df_nan.isnull().sum() / len(df_nan) * 100
```

```python
df = df.drop(["Mechanics", "Domains"], axis=1)
df
```

```python
df = df.dropna(subset=["Year Published"])
df.isnull().sum() / len(df_nan) * 100
```

```python
df = df.drop(["ID", "Name"], axis=1)
df
```

```python
df["Owned Users"] = df["Owned Users"].fillna(df["Owned Users"].mean())
df.describe()
```

```python
from sklearn.preprocessing import MinMaxScaler

X = df.drop(columns=["Rating Average"])
y = df["Rating Average"]

mmsc = MinMaxScaler()
X_norm = mmsc.fit_transform(X)
X_norm = pd.DataFrame(X_norm, columns=X.columns)
X_norm
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X_norm, y, test_size=0.3)
```

```python
from sklearn.linear_model import LinearRegression

model = LinearRegression()

model.fit(X_train, y_train)

y_pred = model.predict(X_test)
```

```python
from sklearn import metrics

print("MAE", metrics.mean_absolute_error(y_test, y_pred))
print("MSE", metrics.mean_squared_error(y_test, y_pred))
print("R2 Score:", model.score(X_test, y_test))
```

```python
df_test = pd.read_csv("test_data.csv")
df_test.info()
```

```python
df_test["Complexity Average"] = pd.to_numeric(
    df_test["Complexity Average"].astype(str).str.replace(',', '.'), 
    errors='coerce'
)
df_test = df_test.drop(["ID", "Name", "Mechanics", "Domains"], axis=1)
df_test["Owned Users"] = df_test["Owned Users"].fillna(df_test["Owned Users"].mean())
df_test.info()
```

```python
X_test_norm = mmsc.transform(df_test)
X_test_norm = pd.DataFrame(X_test_norm, columns=df_test.columns)
X_test_norm
```

```python
y_pred_test = model.predict(X_test_norm)

pd.DataFrame({'Rating Average': y_pred_test}).to_csv('sub.csv', index_label='index')
```

## Списки

Вам дан список. Поменяйте местами первую и вторую половину списка.

```python
arr = list(range(10))
print(arr)
arr = arr[len(arr) // 2:] + arr[:len(arr) // 2]
```

Вам дан список. Поменяйте в нём местами блоки элементов: каждый нечётный со первого по пятый с начала с каждым чётный со второго по пятый с конца. 

```python
arr = list(range(1, 21))
print(arr)
arr[:5:2], arr[-2:-7:-2] = arr[-2:-7:-2], arr[:5:2]
arr
```

## Генераторы

Вам дан список. Создайте новый список, в котором останутся только чётные элементы кратные трём. 

```python
arr = list(range(1, 21))
print(arr)
new_arr = [val for i, val in enumerate(arr) if i % 2 == 0 and val % 3 == 0]
new_arr
```

## NumPy

Дан массив 16х16
Создайте новый массив 4х4 , элементы которого будут суммой блоков 4х4 исходного массива

```python
from numpy.lib.stride_tricks import sliding_window_view

arr = np.arange(16 * 16).reshape(16, 16) % 16
arr[1::2, :] = 0
print(arr)
print(sliding_window_view(arr, (4, 4)).shape)                                                                                                                                                                  
sliding_window_view(arr, (4, 4), )[::4, ::4].sum(axis=(2, 3))
```

Дан массив, вставьте между каждым значением сумму соседних элементов

```python
arr = np.arange(1, 21)
print(arr)
new_arr = arr[1:] + arr[:-1]

print(np.vstack((arr[:-1], new_arr)))

print(np.append(np.vstack((arr[:-1], arr[1:] + arr[:-1])).T.flatten(), arr[-1]))
```

```python
numeric_columns = df.select_dtypes(exclude="object").columns
plt.subplots(3, 4, figsize=(14, 8))
for i, column in enumerate(numeric_columns):
    plt.subplot(3, 4, i + 1)
    df.boxplot(column=column)
```

```python
outliers_df = pd.DataFrame()
for column in df.select_dtypes(exclude="object"):
    q1 = df[column].quantile(0.25)
    q3 = df[column].quantile(0.75)
    iqr = q3 - q1
    outliers_df[column] = df[column].apply(lambda x: x > q3 + 1.5 * iqr or x < q1 - 1.5 * iqr)
print(outliers_df[outliers_df.any(axis=1) == True].index)
```
